### HOW TO READ AND WRITE AN IMAGE 

```python
import cv2
import matplotlib.pyplot as plt
```

#### 0 >>gray image

```python
img_1=cv2.imread('chameleon.webp',0) #0 for gray
# cv2.imshow('image',img_1)
plt.imshow(img_1,cmap='gray')
```

#### 1 >>rgb image

```python
img_2=cv2.imread('chameleon.webp',1) #1 for color
# cv2.imshow('image',img_2)
plt.imshow(img_2,cmap='gray')
plt
```

#### -1 >>alpha channel image

```python
img_3=cv2.imread('chameleon.webp',-1) #-1 for alpha channel
# cv2.imshow('image',img_3)
plt.imshow(img_3,cmap='gray')
```

## Waitkey

```python
cv2.imshow('image',img_1)

k=cv2.waitKey(0)# 0 means that it requires a button to close
if(k==27):# 27 represents 'esc' button
    cv2.destroyAllWindows()
elif(k==ord('s')): # means if we click in 's' button
    cv2.imwrite('Saved_image.webp',img_1) #first parameter (path+.extension)
    cv2.destroyAllWindows()
    
```
