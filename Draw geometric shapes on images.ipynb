# Draw geometric shapes on images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros([512, 512, 3], np.uint8) #create 512*512 black imag with 3 channels

cv2.line(img, (0,0), (255,255), (0, 255, 0), 10) #line takes (imag,startpoint ,endpoint,color thickness)
plt.imshow(img)
plt.title('image with  Drawn Line')
plt.show()
```

```python
img = np.zeros([512, 512, 3], np.uint8) #create 512*512 black imag with 3 channels

img = cv2.arrowedLine(img, (0,255), (255,255), (255, 0, 0), 10) # arrow takes (imag,startpoint ,endpoint,color thickness)
plt.imshow(img)
plt.title('image with  Drawn Arrow')
plt.show()
```

```python
img = np.zeros([512, 512, 3], np.uint8) #create 512*512 black imag with 3 channels

img = cv2.rectangle(img, (384, 0), (510, 128), (0, 0, 255), -1) #rectangle takes (imag,startpoint ,endpoint,color thickness) -1 >>fu
img = cv2.rectangle(img, (10, 10), (150, 150), (0, 255, 0), 3)  
plt.imshow(img)
plt.title('image with  Drawn Rectangles')
plt.show()
```

```python

img = np.zeros([512, 512, 3], np.uint8) #create 512*512 black imag with 3 channels

img = cv2.circle(img, (447, 63), 63, (0, 255, 0), -1) #circle takes (img,center,radius,color ,thichkness) -1>>full of
img = cv2.circle(img, (150, 150), 63, (255, 255, 255), 3)
plt.imshow(img)
plt.title('image with  Drawn Circles')
plt.show()
```

```python

img = np.zeros([512, 512, 3], np.uint8) #create 512*512 black imag with 3 channels
font = cv2.FONT_HERSHEY_SIMPLEX #type of font
img = cv2.putText(img, 'Esraa ', (40, 150), font, 5, (255, 255, 255), 10)# text takes (image,text,postion,font,zoom,colot,thickness)

plt.imshow(img)
plt.title('image with Text')
plt.show()
```

```python

img = np.zeros([512, 512, 3], np.uint8) #create 512*512 black imag with 3 channels

pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32) #make array to draw it by polylines
pts = pts.reshape((-1,1,2))#reshape it
img = cv2.polylines(img,[pts],True,(255,255,255),4) #polyline takes(image,arr postion,bool val,color,thickness)  True>>Completeshape, False >>incomplete shape


plt.imshow(img)
plt.title('image with  Drawn polylines')
plt.show()
```
