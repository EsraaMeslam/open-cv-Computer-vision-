## How to Read, Write, Show Videos from Camera in OpenCV

```python
import cv2
```

### Record RGB video

```python
cap = cv2.VideoCapture(0) # 0 if you want to catpture a video
                          # say ('path/video.mp4') path+name+extension of a video in ur device

fourcc = cv2.VideoWriter_fourcc(*'XVID') #fourcc for resoluation


out = cv2.VideoWriter('Recorderd_vid.avi', fourcc, 20.0, (640, 480)) #(name+extension ) of the recorder video
                                                                     #fourcc for the resoluation
                                                                     # frames per second here 20 frame per second
                                                                     #(width,height)


print(cap.isOpened()) #check if it True or False

while(cap.isOpened()): ##== True
    ret, frame = cap.read()  #ret is bool val
                             #read frames

    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # get width
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) #get height
        
        frame = cv2.flip(frame, 1) # Flip the image horizontally
        
        out.write(frame) #write frames

        cv2.imshow('frame', frame)  
        if cv2.waitKey(1) & 0xFF == ord('q'): #click on button 'q' to stop recording
            break
    else:
        break

cap.release()
out.release()

cv2.destroyAllWindows()
```

### Record Gray video

```python
cap = cv2.VideoCapture(0) # 0 if you want to catpture a video
                          # say ('path/video.mp4') path+name+extension of a video in ur device

fourcc = cv2.VideoWriter_fourcc(*'XVID') #fourcc for resoluation


out = cv2.VideoWriter('Recorderd_vid.avi', fourcc, 20.0, (840, 980)) #(name+extension ) of the recorder video
                                                                     #fourcc for the resoluation
                                                                     # frames per second here 20 frame per second
                                                                     #(width,height)


print(cap.isOpened()) #check if it True or False

while(cap.isOpened()): ##== True
    ret, frame = cap.read()  #ret is bool val
                             #read frames

    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # get width
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) #get height
        
        frame = cv2.flip(frame, 1) # Flip the image horizontally
        
        out.write(frame) #write frames
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
       

        cv2.imshow('frame', gray)  
        if cv2.waitKey(1) & 0xFF == ord('q'): #click on button 'q' to stop recording
            break
    else:
        break

cap.release()
out.release()

cv2.destroyAllWindows()
```
